```python
import pandas as pd
```

```python
data=pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
data
```

```python
x=data.iloc[:,0:4]
```

```python
x
```

```python
y=data.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

```python
transformer=ColumnTransformer(transformers=[('tran1',StandardScaler(),['user ID'])
                                            ('tran2',OneHotEncoder(drop='first',spase=False),['gender']),
                                            ('tran3',StandardScaler,['Age']),
                                            ('tran4',StandardScaler(),['EstimatedSalary'])],reminder='passthrough')
```

```python
transformer = ColumnTransformer(transformers=[
    ('tran1', StandardScaler(), ['User ID']),
    ('tran2', OneHotEncoder(drop='first', sparse=False), ['Gender']),
    ('tran3', StandardScaler(), ['Age']),
    ('tran4', StandardScaler(), ['EstimatedSalary'])], remainder='passthrough')
```

```python
x_train_new=transformer.fit_transform(x_train)
```

```python
x_train=pd.DataFrame(x_train_new,columns=x_train.columns)
```

```python
x_train
```

```python
x_test_new=transformer.fit_transform(x_test)
```

```python
x_test=pd.DataFrame(x_test_new,columns=x_test.columns)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
test=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test,y_test)
```
